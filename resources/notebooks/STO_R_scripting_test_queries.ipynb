### Verifying R Package Version

```
SELECT DISTINCT rscript_version
FROM SCRIPT (
    SCRIPT_COMMAND('Rscript --vanilla -e "write(R.version.string, stdout())"')
    RETURNS('rscript_version VARCHAR(50)')
);
```

### Printing the Value of Pi

```
SELECT DISTINCT piVal
FROM SCRIPT (
  SCRIPT_COMMAND('Rscript --vanilla -e "write(pi, stdout())"')
  RETURNS('piVal VARCHAR(20)')
);
```

### Generate a Random Number in R and Return It

```
SELECT DISTINCT random_value
FROM SCRIPT (
  SCRIPT_COMMAND('Rscript --vanilla -e "write(runif(1, 0, 100), stdout())"')
  RETURNS('random_value FLOAT')
);
```

### Running a Simple Calculation in R

```
SELECT DISTINCT sqrt_value
FROM SCRIPT (
  SCRIPT_COMMAND('Rscript --vanilla -e "write(sqrt(625), stdout())"')
  RETURNS('sqrt_value FLOAT')
);
```
